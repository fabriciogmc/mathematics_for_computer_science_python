```python
# Exemplo de multiplicação de números complexos e visualização
# no plano cartesiano
#
# Autor:  Fabrício Galende Marques de Carvalho

import matplotlib.pyplot as plt
import numpy as np

z = 1 + 2j
x = z.real
y = z.imag
plt.figure()
plt.scatter(x, y, color='black') # plota o ponto

plt.plot([0, x], [0, y], linestyle='--', color ='black')
plt.xlim(1.1*-np.abs(x), 1.1*np.abs(x))

# ajuste visual
plt.axhline(0, color='black')  # eixo real
plt.axvline(0, color='black')  # eixo imaginário
plt.grid()

plt.xlabel("Parte Real")
plt.ylabel("Parte Imaginária")
plt.title("Número complexo no plano.")
plt.show()
```

```python
from math import sqrt
#z_mult =  complex(sqrt(2)/2, sqrt(2)/2)
z_mult =  complex(0,1) 
zr = z_mult * z

x = zr.real
y = zr.imag
plt.figure()
plt.scatter(x, y, color="black")

plt.plot([0, x], [0, y], linestyle='--', color = "black")
plt.xlim(1.1*-np.abs(x), 1.1*np.abs(x))

plt.axhline(0, color="black")  
plt.axvline(0, color="black")  
plt.grid()

plt.xlabel("Parte Real")
plt.ylabel("Parte Imaginária")
plt.title("Número complexo no plano, rotacionado em 90 graus.")
plt.show()
```
