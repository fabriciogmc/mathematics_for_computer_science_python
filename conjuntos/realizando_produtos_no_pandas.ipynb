```python
# Exemplo de operações entre conjuntos aplicados data frames
# criados a partir de dicionários.
# Exemplo de produto entre conjuntos com Pandas
#
# Autor: Fabrício Galende Marques de Carvalho

import pandas as pd
from pandas import read_csv

#Criando um dataframe a partir de lista de valores em dicionários:
nomes = ["fabrício", "jane", "thales", "eduardo", "fabíola"]
cidades = ["são josé", "taubaté", "itatiaia", "taubaté", "itatiaia" ]
dataframe_1 = pd.DataFrame({"nome": nomes})
dataframe_2 = pd.DataFrame({"cidade": cidades})

print(dataframe_1)
print("\n")
print(dataframe_2)
print("\n")
```

```python
# Realizando o produto cartesiano com merge, crossjoin
dataframe_prod = dataframe_1.merge(dataframe_2, how="cross")
print(dataframe_prod)
```
