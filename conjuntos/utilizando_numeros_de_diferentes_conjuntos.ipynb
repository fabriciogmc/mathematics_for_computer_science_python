```python
# Casos especiais de conjuntos numéricos
#
# Autor: Fabrício Galende Marques de Carvalho

# Utilizando números naturais

natural_1 = 2
natural_2 = 20
natural_3 = natural_1 + natural_2
natural_4 = natural_1 * natural_2
print(natural_3)
print(natural_4)
```

```python
#Utililzando números inteiros
inteiro_1 = 2
inteiro_2 = -1
inteiro_3 = inteiro_2 - inteiro_1
print(inteiro_3)
```

```python
# Utilizando números racionais
racional_1 = 1.0/3.0
racional_2 = -2.0/3.0
racional_3 = racional_1 + racional_2
print(racional_3)
print("\n")

# Comparação incorreta de números racionais:
racional_4 = 0.1
racional_5 = 0.4-0.3
print(f"{racional_4} == {racional_5}", racional_4 == racional_5)
print("\n")

# Comparação correta de números racionais
tolerancia = 0.0001
print(f"abs({racional_4} - {racional_5}) < {tolerancia}" , abs(racional_4-racional_5) < tolerancia)
print("\n")
```

```python
# Utilizando números iracionais:
from math import sqrt, pi
irracional_1 = sqrt(2)
irracional_2 = pi
print(irracional_1)
print(irracional_2)
```

```python
from matplotlib import pyplot as plt

# Utilizando números complexos
complexo_1 = 1 + 2j
x = complexo_1.real
y = complexo_1.imag

plt.scatter(x,y, color="blue")
plt.arrow(0,0, x,y, head_width=0.03, length_includes_head = True, color="black")
plt.xlabel("Parte real")
plt.ylabel("Parte imaginária")
plt.grid()
plt.show()
```
