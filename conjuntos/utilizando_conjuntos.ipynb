```python
# Criando e utilizando conjuntos
#
# Autor: Fabrício Galende Marques de Carvalho


# Criando um conjunto numérico a partir da classe set

meu_conjunto_numerico = set()
meu_conjunto_numerico.add(1)
meu_conjunto_numerico.add(2)
meu_conjunto_numerico.add(1)
print(meu_conjunto_numerico)
```

```python
# Verificando se um elemento pertence ao conjunto de números e removendo esse elemento:
if 1 in meu_conjunto_numerico:
    meu_conjunto_numerico.remove(1)
print("Conjunto numérico após a remoção de 1: ", meu_conjunto_numerico)
```

```python
# Criando um conjunto utilizando {}

meu_conjunto_de_nomes = {"Fabricio", "Jane", "Jane"}
print(meu_conjunto_de_nomes)
```

```python
# Iterando sobre elementos de um conjunto:
for nome in meu_conjunto_de_nomes:
    print(nome)
```

```python
# Removendo um elemento do conjunto utilizando a notação {}
meu_conjunto_de_nomes.remove("Fabricio")
print(meu_conjunto_de_nomes)
```

```python
# Criando um conjunto com tipos diferentes
meu_conjunto_varios_tipos = set(["Thales", 1])
print(meu_conjunto_varios_tipos)
```
