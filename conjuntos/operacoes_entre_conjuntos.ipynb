```python
# Operações entre conjuntos:
#
# Autor: Fabrício Galende Marques de Carvalho

from itertools import product


conjunto_1 = set([1, 2, 3, 4])
conjunto_2 = set([3,4])
conjunto_3 = set([2, 1, 3, 4])

print("Verificando se conjunto_1 é igual ao conjunto_2")
print(conjunto_1 == conjunto_2)
print("\n")

print("Verificando se conjunto_1 é igual ao conjunto_3")
print(conjunto_1 == conjunto_3)
print("\n")

print("Verificando se conjunto_2 é subconjunto de conjunto_1 : ")
print(conjunto_2.issubset(conjunto_1))
print("\n")

print("Verificando se conjunto_1 é subconjunto de conjunto_2 : ")
print(conjunto_1.issubset(conjunto_2))
print("\n")

print("Calculando a união entre os conjuntos: ")
print( conjunto_1.union(conjunto_2) )
print( conjunto_1 | conjunto_2 )
print("\n")

print("Calculando a intersecção entre os conjuntos: ")
print( conjunto_1.intersection(conjunto_2) )
print( conjunto_1 & conjunto_2 )
print("\n")

print("Calculando a diferença entre os conjuntos: ")
print(conjunto_1 - conjunto_2)
print("\n")

print("Calculando o produto entre dois conjuntos: ")
produto_1_2 = product(conjunto_1, conjunto_2)
print("Produto conjunto 1 x conjunto 2: ", list(produto_1_2))

```
