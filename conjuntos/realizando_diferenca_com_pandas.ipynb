```python
# Exemplo de operações entre conjuntos aplicados a tabelas
# armazenadas em arquivos .CSV. 
# Exemplo de diferença entre conjuntos com Pandas

import pandas as pd
from pandas import read_csv

#Criando um dataframe a partir de lista de valores em dicionários:
nomes = ["fabrício", "jane", "thales", "eduardo", "fabíola"]
cidades = ["são josé", "taubaté", "itatiaia", "taubaté", "itatiaia" ]
dataframe = pd.DataFrame({"nome":nomes, "cidade":cidades})


#Examinando o conteúdo do DataFrame
print(dataframe)
print(dataframe.columns)
print("\n")

# Contando quantas linhas e colunas há no dataframe:
print("Número de linhas: ", dataframe.shape[0])
print("Número de colunas: ", dataframe.shape[1])
print("\n")

#Examinando uma linha pelo índice:
print(dataframe.iloc[0])
print(dataframe.iloc[0].nome)
print("\n")

#Selecionando e examinando outro índice
dataframe_index_nome = dataframe.set_index("nome")
print(dataframe_index_nome)
print("\n")
print(dataframe_index_nome.loc["fabrício"].cidade)


```

```python
# Realizando a primeira consulta:
dataframe_1 = dataframe[dataframe["nome"] > "fabrício"]
print("Resultado da primeira consulta: ")
print(dataframe_1)
print("\n")

# Realizando a segunda consulta
print("Resultado da segunda consulta: ")
dataframe_2 = dataframe[dataframe["nome"] > "eduardo"]
print(dataframe_2)
print("\n")


# Realizando a diferença entre as duas consultas
# leftjoin criando coluna indicadora de ocorrência
dataframe_diff = dataframe_2.merge(dataframe_1.drop_duplicates(), how='left', indicator=True) 
print(dataframe_diff)
print("\n")

dataframe_diff = dataframe_diff.query('_merge == "left_only"')  #Filtrando somente as que estão na esquerda
dataframe_diff = dataframe_diff.drop(columns=['_merge']) # eliminando a coluna indicadora de ocorrência.
print(dataframe_diff)
```
