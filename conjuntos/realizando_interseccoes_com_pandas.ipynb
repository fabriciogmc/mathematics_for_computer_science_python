```python
# Exemplo de operações entre conjuntos aplicados a tabelas
# armazenadas em arquivos .CSV. 
# Exemplo de intersecção (join) com pandas

import pandas as pd
from pandas import read_csv

#Carregando os dados do arquivo de usuários
dataframe_u = read_csv("dados_usuario.csv", skipinitialspace = True)

#Carregando os dados do arquivo de cidades
dataframe_c = read_csv("dados_cidade.csv", skipinitialspace = True)


#Examinando o conteúdo dos DataFrames
print(dataframe_u)
print(dataframe_u.columns)
print("\n")

print(dataframe_c)
print(dataframe_c.columns)
print("\n")
```

```python
# Realizando o inner join selecionando os índices correspondentes.
dataframe_join = dataframe_u.set_index("cidade").join(dataframe_c.set_index("nome"), how="inner")

print(dataframe_join)
```

```python
# Realizando o inner join selecionando os índices correspondentes e preservando o índice do dataframe original.
dataframe_join = dataframe_u.join(dataframe_c.set_index("nome"), on="cidade", how="inner")

print(dataframe_join)
```

```python
# Realizando o left join selecionando os índices correspondentes.
dataframe_join = dataframe_u.set_index("cidade").join(dataframe_c.set_index("nome"), how="left")

print(dataframe_join)
```

```python
# Realizando o right join selecionando os índices correspondentes.
dataframe_join = dataframe_u.set_index("cidade").join(dataframe_c.set_index("nome"), how="right")

print(dataframe_join)
```

```python
# Realizando o full outer join selecionando os índices correspondentes.
dataframe_join = dataframe_u.set_index("cidade").join(dataframe_c.set_index("nome"), how="outer")

print(dataframe_join)
```

```python
# Realizando o inner join utilizando o merge
dataframe_join = dataframe_u.merge(dataframe_c, left_on="cidade", right_on="nome",how="inner", suffixes=["_u", "_c"])
print(dataframe_join)
```

```python
# Realizando o left join utilizando o merge
dataframe_join = dataframe_u.merge(dataframe_c, left_on="cidade", right_on="nome",how="left", suffixes=["_u", "_c"])
print(dataframe_join)
```

```python
# Realizando o right join utilizando o merge
dataframe_join = dataframe_u.merge(dataframe_c, left_on="cidade", right_on="nome",how="right", suffixes=["_u", "_c"])
print(dataframe_join)
```

```python
# Realizando o full outer join utilizando o merge
dataframe_join = dataframe_u.merge(dataframe_c, left_on="cidade", right_on="nome",how="outer", suffixes=["_u", "_c"])
print(dataframe_join)
```
