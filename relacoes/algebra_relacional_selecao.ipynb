```python
# Aplicação da álgebra relacional às operações
# sobre dados tabulares: Operação de seleção
#
# Autor: Fabrício Galende Marques de Carvalho

from pandas import DataFrame

nome = ["Ronaldo", "Beatriz", "Jane"]
genero = ["masculino", "feminino", "feminino"]

usuario_df = DataFrame({"nome":nome, "genero":genero})

print("Visualizando a tabela de usuários armazenada no dataframe:")
print(usuario_df)

```

```python
# Realizando uma seleção por igualdade de valor de atributo (sigma_(atrib=x)(usuario_df) ):

select_df = usuario_df[usuario_df["nome"]=="Ronaldo"]

print(select_df)
```

```python
# Realizando uma seleção por faixa de valores de atributo ( sigma_(atrib != x)(usuario_df) )

select_df = usuario_df[usuario_df["nome"]!="Ronaldo"]

print(select_df)
```
