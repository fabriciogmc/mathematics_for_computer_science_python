```python
# Aplicação da álgebra relacional às operações
# sobre dados tabulares: Operação de projeção.
#
# Autor: Fabrício Galende Marques de Carvalho

from pandas import DataFrame

nome = ["Ronaldo", "Beatriz", "Jane"]
genero = ["masculino", "feminino", "feminino"]
data_nasc = ["01/01/1960", "01/01/1970", "01/01/1980"]

usuario_df = DataFrame({"nome":nome, "genero":genero, "data_nasc":data_nasc})

print("Visualizando a tabela de usuários armazenada no dataframe:")
print(usuario_df)

```

```python
# Realizando a projeção de um único atributo (pi_(atrib_x)(usuario_df) ):

project_1_df = usuario_df[["nome"]]

print(project_1_df)
```

```python
# Realizando projeção de uma lista de atributos ( pi_(atrib_x, a)(usuario_df) )

select_df = usuario_df[usuario_df["nome"]!="Ronaldo"]

print(select_df)
```
