```python
"""
Representando funções através de tabelas, diagramas de 
dispersão e gráficos.

Autor: Fabrício Galende Marques de Carvalho

"""

import pandas as pd
from matplotlib import pyplot as plt

x = [1, 2, 3]
y = [3, 5, 7]

tabela_funcao = pd.DataFrame({"x":x, "y":y})
```

```python
# Visualizando função através de dados tabulares
print(tabela_funcao)
```

```python
#Visualizando um diagrama de dispersão
plt.scatter(tabela_funcao["x"].to_numpy(), tabela_funcao["y"].to_numpy(), marker="o", s=100)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
# Visualizando através de um gráfico
plt.plot(tabela_funcao["x"].to_numpy(), tabela_funcao["y"].to_numpy(), color="black")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
