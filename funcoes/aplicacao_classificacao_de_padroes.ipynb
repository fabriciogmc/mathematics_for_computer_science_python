```python
"""
Neste exemplo, mostra-se como a função distância euclidiana, que
é uma função algébrica, pode ser utilizada na classificação
de uma imagem monocromática

Autor: Fabrício Galende Marques de Carvalho

"""

from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

# Leitura e redimensionamento dos arquivos de imagem

img_mulher = Image.open("data/mulher_foto_01.bmp")
img_homem = Image.open("data/homem_foto_01.bmp")
img_desconhecido = Image.open("data/nao_classificado.bmp")

largura = 100
altura = 130

img_mulher = img_mulher.resize((largura,altura))
img_homem = img_homem.resize((largura,altura))
img_desconhecido = img_desconhecido.resize((largura,altura))


```

```python
# Visualizando metadados das imagens e as imagens em si

# Metadados
print(img_mulher.size)
print(img_homem.size)
print(img_desconhecido.size)

#Dados
plt.figure()
plt.imshow(img_mulher, cmap='gray')
plt.title("Imagem de mulher")
plt.figure()
plt.imshow(img_homem, cmap='gray')
plt.title("Imagem de homem")
plt.figure()
plt.imshow(img_desconhecido, cmap='gray')
plt.title("Imagem não classificada (desconhecida)")


```

```python
# Convertendo as imagens para arrays bidimensionais numpy
img_mulher_array_2d = np.array(img_mulher)
print("Array 2d correspondente (mulher): ")
print(img_mulher_array_2d)
img_mulher_array_1d = img_mulher_array_2d.flatten()
print("\nArray 'achatado' (mulher): ")
print(img_mulher_array_1d)


img_homem_array_1d = np.array(img_homem).flatten()
img_desconhecido_array_1d = np.array(img_desconhecido).flatten()

```
